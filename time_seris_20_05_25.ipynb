```python
import pandas as pd
```

```python
df= pd.read_csv("C:\\Users\\Prathamesh\\Desktop\\time_series_python\\Time seriesData\\starbucks.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

df.sort_index(inplace=True)

print(df.head())
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

weights = pd.Series([1, 2, 3, 4, 5])

def weighted_moving_average(series, weights):
    return series.rolling(window=len(weights)).apply(lambda x: sum(weights * x), raw=True)

df['WMA_5'] = weighted_moving_average(df['Close'], weights)

plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='Close Price', alpha=0.5)
plt.plot(df['WMA_5'], label='5-Day WMA', color='yellow')
plt.title('Close Price vs 5-Day Weighted Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')

plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
df['EMA_10'] = df['Close'].ewm(span=10, adjust=False).mean()

plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='Close Price', color='blue', alpha=0.5)
plt.plot(df['EMA_10'], label='10-Day EMA', color='red')
plt.title('Close Price vs 10-Day Exponential Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.tight_layout()

plt.show()
```
